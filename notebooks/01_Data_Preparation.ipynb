```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)

print("Libraries imported successfully.")
```

```python
# Load Superstore Sales Dataset

df = pd.read_csv("../data/raw/train.csv")

print("Dataset loaded successfully.")
```

```python
# Load Video Game Sales Dataset

vg_df = pd.read_csv("../data/raw/vgsales.csv")

print("Video Game Sales dataset loaded successfully.")
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], format="%d/%m/%Y")

df["Ship Date"] = pd.to_datetime(df["Ship Date"], format="%d/%m/%Y")

print("Data columns converted successfully")
```

```python
df[["Order Date", "Ship Date"]].head()
```

```python
df["Year"] = df["Order Date"].dt.year

df["Month"] = df["Order Date"].dt.month

df["Month Name"] = df["Order Date"].dt.month_name()

df["Quarter"] = df["Order Date"].dt.quarter

df["Week"] = df["Order Date"].dt.isocalendar().week

df["Day"] = df["Order Date"].dt.day

df["Day Name"] = df["Order Date"].dt.day_name()

print("Time-based features created successfully")
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer" 
    else:
        return "Autumn"

df["Season"] = df["Month"].apply(get_season)

df[["Order Date", "Season"]].head()
```

```python
df.to_csv(
    "../data/processed/superstore_sales_processed.csv", 
    index=False
)

print("Processed dataset saved successfully")
```

```python
vg_df.to_csv(
    "../data/processed/vgsales_processed.csv", 
    index=False
)
print("Video Game dataset saved successfully")
```
